### Okay, this data is pretty simple and straightforward. Let's clean it up, give it a trim and create some visuals. ###

```python
# First, it would be helpful to have the general info on the data set as reference

coolwarmcenter.info()
```

***Again, this gives us important column and data entry information for the data set. This will be useful for step below***

```python
# Now I want to look at a dataframe of a subset of only the columns I think would be useful to pull into an asset map

coolwarmcenter[['X','Y','source','cat1','cat2','org_name','Name','addrln1','addrln2','city','hours','phones','url','description','zip','link','latitude','longitude','email','date_updated','dis_status']]
```

***I've trimmed out all excessive columns and created a dataframe that includes only the columns I specified above. Next, I will create a new variable to house my trimmed list***

```python
# create a new variable to house my trimmed list

coolwarmcenter_trimmed = coolwarmcenter[['X','Y','source','cat1','cat2','org_name','Name','addrln1','addrln2','city','hours','phones','url','description','zip','link','email','date_updated','dis_status']].copy()
coolwarmcenter_trimmed
```

***Now we have a trimmed dataframe under new variable 'coolwarmcenter_trimmed' which I'll use to create a visual***

### Unfortunately, since this is not a shape file we cannot use geopandas to map points which is what I ultimately want to do. While I don't have time for this assignment, there seems to be a workaround for plotting points for a .csv file. Instead, I'll visualize how the cities compare in terms of number of cooling centers ### 

**The workaround involves creating a concatonated 'POINT' field using 'X' and 'Y' variables (aka lat/long) and converting the data type to 'geometry' so it can then be plotted using geopandas in Python.**

```python
# I'm interested in viewing information for only the cooling centers in the city of LA. Will query for this.

coolwarmcenter_trimmed[coolwarmcenter_trimmed.city == 'Los Angeles']
```

***The output dataframe is showing me the trimmed column information for ONLY the cooling centers located in the city of LA. I'll create a new variable for this LA-specific trimmed dataframe***

```python
# Going to create an even further trimmed dataframe limited to cooling centers in the city of LA

coolwarmcenter_trimmed_LA = coolwarmcenter_trimmed[coolwarmcenter_trimmed.city == 'Los Angeles']
coolwarmcenter_trimmed_LA
```

***Same trimmed dataframe as before, just limited to city of LA cooling centers***

```python
#In order to visually compare the number of cooling centers in each zip code within the city of LA, I'll need to create a dataframe for value_counts of 'zip'

coolwarmcenter_trimmed_LA['zip'].value_counts()
LA_zipcounts = coolwarmcenter_trimmed_LA['zip'].value_counts()
LA_zipcounts
```

***Output gives us the number of centers located in each zip code within city of LA***

```python
type(LA_zipcounts)
```

***To be able to work with the data in this table I'll need to convert to a dataframe type***

```python
# convert series to dataframe

LA_zipcounts = LA_zipcounts.reset_index()
LA_zipcounts
```

```python
# series should now be a dataframe. Just going to double check

type(LA_zipcounts)
```

***series is now dataframe***

```python
# look at current column names in order to rename

LA_zipcounts.columns.to_list()
```

***existing column names***

```python
# rename columns

LA_zipcounts.columns = ['zip', 'count']
LA_zipcounts
```

***successfully renamed columns***

### Now let's plot what we've got! ###

```python
# Generate bar graph with additional arguments (to make it look nicer)

LA_zipcounts.plot.bar(x = 'zip', y = 'count', title = 'Number of cooling centers in city of LA by zip code')
```
